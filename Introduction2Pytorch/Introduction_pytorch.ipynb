## Autograd and Gradients

Autograd is PyTorch’s automatic differentiation engine, which is crucial for training neural networks. It allows you to automatically compute gradients, a key step in the backpropagation process.

- **Automatic Differentiation**:
When you perform operations on tensors that have requires_grad=True, PyTorch builds a dynamic computation graph. This graph records all the operations performed on the tensor, enabling PyTorch to compute gradients automatically.

- **Backward Pass**:
Once you have a scalar loss computed from your model’s outputs, calling .backward() on that loss triggers the backward pass. This computes the gradients of the loss with respect to every tensor that has requires_grad=True and stores these gradients in the tensor’s .grad attribute.

````python
import torch

# Create a tensor with gradient tracking enabled
x = torch.tensor([2.0, 3.0], requires_grad=True)

# Define a simple function: y = x^2 + 3x
y = x**2 + 3 * x

# Compute a scalar loss by summing the elements of y
loss = y.sum()
print("Loss:", loss.item())

# Perform backpropagation to compute gradients of loss with respect to x
loss.backward()

# The computed gradients are stored in x.grad
print("Gradients for x:", x.grad)
````

````python
import torch

# Create a tensor with gradient tracking enabled
x = torch.tensor([2.0, 3.0], requires_grad=True)
loss_list=[]
# Perform multiple iterations
for i in range(5):  # Run for 5 iterations
    # Define a simple function: y = x^2 + 3x
    y = x**2 + 3 * x
    
    # Compute a scalar loss by summing the elements of y
    loss = y.sum()
    loss_list.append(loss)
    print(f"Iteration {i+1}, Loss:", loss.item())
    
    # Zero out the gradients from previous iteration
    if x.grad is not None:
        x.grad.zero_()
    
    # Perform backpropagation to compute gradients of loss with respect to x
    loss.backward()
    
    # The computed gradients are stored in x.grad
    print(f"Iteration {i+1}, Gradients for x:", x.grad)
    
    # Update x using gradient descent (as an example)
    with torch.no_grad():
        x -= 0.1 * x.grad  # Simple gradient descent with learning rate 0.1
    
    print(f"Iteration {i+1}, Updated x:", x)
    print("-" * 50)
````

## **2. Building Models and Training Cycle (40-50 minutes)**

### **Defining Models with `nn.Module`**
# In PyTorch, neural network models are built using the `torch.nn.Module` class. This class allows encapsulating the model architecture and inference logic in the `forward()` method.
#
# #### **Example: Building an MLP for Classification**



````python

import torch
import torch.nn as nn
import torch.nn.functional as F

class MLP(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, output_size)
    
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
````

### Complete example

````python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Define the neural network architecture
class MNISTNet(nn.Module):
    def __init__(self):
        super(MNISTNet, self).__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, 10)
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Load and preprocess dataset
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])  # More standard normalization
])

# Download training data
train_dataset = datasets.MNIST(root='data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='data', train=False, download=True, transform=transform)

# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# Initialize the model
model = MNISTNet().to(device)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
num_epochs = 5
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        
        # Zero the gradients
        optimizer.zero_grad()
        
        # Forward pass
        output = model(data)
        loss = criterion(output, target)
        
        # Backward pass and optimization
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        if batch_idx % 100 == 99:
            print(f'Epoch [{epoch + 1}/{num_epochs}], Batch [{batch_idx + 1}/{len(train_loader)}], '
                  f'Loss: {running_loss/100:.4f}')
            running_loss = 0.0

# Evaluation
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for data, target in test_loader:
        data, target = data.to(device), target.to(device)
        output = model(data)
        _, predicted = torch.max(output.data, 1)
        total += target.size(0)
        correct += (predicted == target).sum().item()

print(f'Accuracy on test set: {100 * correct / total:.2f}%')
````

 ## Introduction to Convolutional Neural Networks (ConvNets) in PyTorch
 Convolutional Neural Networks (CNNs or ConvNets) are a specialized type of neural network designed for processing grid-like data, such as images. They are particularly effective for:
 - Image recognition and classification
 - Object detection
 - Computer vision tasks
 - Pattern recognition
 Key components of ConvNets:
 1. Convolutional layers - Extract features using learnable filters
 2. Pooling layers - Reduce spatial dimensions and computation
 3. Fully connected layers - Final classification based on extracted features
 In this section, we'll implement a CNN using PyTorch to improve our MNIST digit classification performance.

````python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Configuración del dispositivo
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Definición de la arquitectura de la red
class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, 3, padding=1)  # 1 canal de entrada, 32 filtros
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(64 * 7 * 7, 128)  # Cálculo del tamaño después de los pooling
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(128, 10)  # 10 clases de salida

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))  # (BS, 32, 14, 14)
        x = self.pool(F.relu(self.conv2(x)))  # (BS, 64, 7, 7)
        x = x.view(-1, 64 * 7 * 7)  # Aplanar para las capas fully connected
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

# Hiperparámetros
BATCH_SIZE = 64
LEARNING_RATE = 0.001
EPOCHS = 10

# Transformaciones y carga de datos
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

train_dataset = datasets.MNIST('./data', 
                               train=True, 
                               download=True, 
                               transform=transform)
test_dataset = datasets.MNIST('./data', 
                              train=False, 
                              transform=transform)

train_loader = DataLoader(train_dataset, 
                          batch_size=BATCH_SIZE, 
                          shuffle=True)
test_loader = DataLoader(test_dataset, 
                         batch_size=BATCH_SIZE, 
                         shuffle=False)

# Inicialización del modelo, loss y optimizer
model = ConvNet().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

# Entrenamiento
for epoch in range(EPOCHS):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        
        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)
        
        # Backward pass y optimización
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        # Estadísticas
        running_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
    
    # Validación
    model.eval()
    test_loss = 0.0
    test_correct = 0
    test_total = 0
    
    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            test_loss += criterion(outputs, labels).item()
            _, predicted = torch.max(outputs.data, 1)
            test_total += labels.size(0)
            test_correct += (predicted == labels).sum().item()
    
    # Impresión de resultados
    print(f"Epoch [{epoch+1}/{EPOCHS}]")
    print(f"Train Loss: {running_loss/len(train_loader):.4f}, Accuracy: {100*correct/total:.2f}%")
    print(f"Test Loss: {test_loss/len(test_loader):.4f}, Accuracy: {100*test_correct/test_total:.2f}%\n")

print("Entrenamiento completado!")

# Evaluación final
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f'Precisión final en test: {100 * correct / total:.2f}%')
````

## Introduction to Recurrent Neural Networks

Recurrent Neural Networks (RNNs) are a powerful class of neural networks specifically designed to process sequential data. Unlike traditional feedforward networks that assume independence between inputs, RNNs have loops in their architecture, allowing information to persist and be used across time steps. This built-in "memory" makes them particularly well-suited for tasks where context and order are important.

### Key Concepts

- **Sequential Data Processing:**  
  RNNs process input sequences one element at a time. At each time step, the network takes the current input along with the hidden state from the previous time step to produce a new hidden state. This process enables RNNs to capture temporal dependencies in data, such as those found in text, audio, and time series.

- **Hidden State and Memory:**  
  The hidden state in an RNN serves as a memory that retains information about previous inputs. This allows the network to build an internal representation of the sequence it has seen so far, which is essential for tasks like language modeling or sequence prediction.

- **Backpropagation Through Time (BPTT):**  
  To train an RNN, the network's parameters are adjusted using a variant of the backpropagation algorithm known as Backpropagation Through Time. BPTT unrolls the network through the sequence's time steps and computes gradients for each step, enabling the network to learn from the entire sequence.

### Applications of RNNs

RNNs have been successfully applied in various domains, including:
- **Natural Language Processing (NLP):** For tasks such as language modeling, machine translation, and text generation.
- **Speech Recognition:** To process audio signals and convert them into text.
- **Time Series Forecasting:** To predict future values in sequences like stock prices or weather data.
- **Video Analysis:** Where sequential frames can be processed to understand temporal dynamics.

### Challenges and Variants

While RNNs are effective at capturing sequential dependencies, they can face challenges when dealing with long sequences due to issues like vanishing and exploding gradients. To mitigate these problems, several variants have been developed:
- **Long Short-Term Memory (LSTM):** Introduces gating mechanisms to better manage the flow of information, enabling the network to learn longer-range dependencies.
- **Gated Recurrent Unit (GRU):** Similar to LSTMs but with a simpler structure, GRUs also help alleviate issues with long-term dependency learning.

Understanding RNNs is fundamental for anyone working with sequential data. As you explore more advanced topics, you’ll see how these networks form the backbone of many modern deep learning applications in language, audio, and time-series analysis.

````python
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import requests
import os

# Download Shakespeare text if not already present
if not os.path.exists('shakespeare.txt'):
    url = 'https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt'
    r = requests.get(url)
    with open('shakespeare.txt', 'w') as f:
        f.write(r.text)

# Read the text
with open('shakespeare.txt', 'r') as f:
    text = f.read()

# Create character mappings
chars = sorted(list(set(text)))
char_to_idx = {ch: i for i, ch in enumerate(chars)}
idx_to_char = {i: ch for i, ch in enumerate(chars)}
n_chars = len(chars)

# Prepare sequences
seq_length = 100
X = []
y = []

for i in range(0, len(text) - seq_length):
    sequence = text[i:i + seq_length]
    label = text[i + seq_length]
    X.append([char_to_idx[char] for char in sequence])
    y.append(char_to_idx[label])

X = torch.tensor(X)
y = torch.tensor(y)

# Create dataset and dataloader
class TextDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

dataset = TextDataset(X, y)
dataloader = DataLoader(dataset, batch_size=128, shuffle=True)

# Define the RNN model
class SimpleRNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleRNN, self).__init__()
        self.hidden_size = hidden_size
        self.rnn = nn.RNN(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)
    
    def forward(self, x, hidden=None):
        batch_size = x.size(0)
        if hidden is None:
            hidden = torch.zeros(1, batch_size, self.hidden_size).to(x.device)
        
        # Convert to one-hot encoding
        x = torch.nn.functional.one_hot(x, num_classes=n_chars).float()
        
        out, hidden = self.rnn(x, hidden)
        out = self.fc(out[:, -1, :])  # Take only the last output
        return out, hidden

# Initialize model, loss function and optimizer
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = SimpleRNN(n_chars, 128, n_chars).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters())

# Training loop
n_epochs = 20
print("Training the model...")

for epoch in range(n_epochs):
    model.train()
    total_loss = 0
    for batch_X, batch_y in dataloader:
        batch_X, batch_y = batch_X.to(device), batch_y.to(device)
        
        optimizer.zero_grad()
        output, _ = model(batch_X)
        loss = criterion(output, batch_y)
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
    
    if (epoch + 1) % 5 == 0:
        print(f'Epoch [{epoch+1}/{n_epochs}], Loss: {total_loss/len(dataloader):.4f}')

# Generate some text
print("\nGenerating sample text...")
model.eval()

# Start with a random sequence from the dataset
start_idx = torch.randint(0, len(dataset), (1,)).item()
current_seq = X[start_idx].unsqueeze(0).to(device)
generated_text = ''.join([idx_to_char[idx.item()] for idx in current_seq[0]])

# Generate 200 new characters
with torch.no_grad():
    for _ in range(200):
        output, _ = model(current_seq)
        prob = torch.softmax(output, dim=1)
        char_idx = torch.multinomial(prob, 1).item()
        
        generated_text += idx_to_char[char_idx]
        
        # Update sequence
        current_seq = torch.cat([current_seq[:, 1:], torch.tensor([[char_idx]]).to(device)], dim=1)

print("Sample generated text:")
print(generated_text)
````

````python
# Generate text from custom input
print("\nGenerating text from custom input...")
model.eval()

# Define custom starting text
custom_text = "Hello Darkness My Old Friend"  # You can modify this text
input_sequence = torch.tensor([[char_to_idx[c] for c in custom_text]]).to(device)

# Initialize generated text with the custom input
generated_text = custom_text

# Generate 200 new characters
with torch.no_grad():
    for _ in range(200):
        output, _ = model(input_sequence)
        prob = torch.softmax(output, dim=1)
        char_idx = torch.multinomial(prob, 1).item()
        
        generated_text += idx_to_char[char_idx]
        
        # Update sequence
        input_sequence = torch.cat([input_sequence[:, 1:], torch.tensor([[char_idx]]).to(device)], dim=1)

print("Input text:", custom_text)
print("\nGenerated continuation:")
print(generated_text)

````

````python
# Save the complete model (architecture and weights)
torch.save({
    'model_state_dict': model.state_dict(),
    'model_architecture': model,
    'char_to_idx': char_to_idx,
    'idx_to_char': idx_to_char,
    'hidden_size': model.hidden_size,  # Save architecture parameters
    'vocab_size': len(char_to_idx)
}, 'text_generation_model_complete.pt')

print("Model architecture and weights saved successfully!")

````
